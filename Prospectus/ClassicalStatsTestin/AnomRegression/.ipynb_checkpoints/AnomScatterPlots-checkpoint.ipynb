```python
##10 day lag
w10 = np.empty((40,561))
e10 = np.empty((40,561))
s10 = np.empty((40,561))
r10 = np.empty((40,561))
ct10 = np.empty((40,561))
cn10 = np.empty((40,561))
g10 = np.empty((40,561))
eh10 = np.empty((40,561))

t10 = np.empty((40,561))

for i in range(0,40,1):
    w10[i,:] = wind10[i,:561]
    e10[i,:] = ephi10[i,:561]
    s10[i,:] = size10[i,:561]
    r10[i,:] = rat10[i,:561]
    ct10[i,:] = cenlat10[i,:561]
    cn10[i,:] = cenlon10[i,:561]
    g10[i,:] = gp[i,:561]
    eh10[i,:] = eh[i,:561]
    
    t10[i,:] = tp[i,40:]
```

```python
##quantiles of anomalies !! 
mx = np.nanquantile(t_gr,1,axis = 0) #maximum
q90 = np.nanquantile(t_gr,0.9,axis = 0) #ninety
q75 = np.nanquantile(t_gr,0.75,axis = 0) #seventy
q50 = np.nanquantile(t_gr,0.5,axis = 0) 
q25 = np.nanquantile(t_gr,0.25,axis = 0)
q10 = np.nanquantile(t_gr,0.1,axis = 0) 
mn = np.nanquantile(t_gr,0,axis = 0) #minimum
```

## Mix of both extreme quantiles in metrics and temperature. 

```python
w_mn = []
tw = []

e_mn = []
te = []

r_mn = []
tr = []

s_mn = []
ts = []

ct_mn = []
tct = []

cn_mn = []
tcn = []

g_mn = []
tg = []

eh_mn = []
teh = []

for i in range(len(t_gr)):
    if t_gr[i] >= np.nanquantile(t_gr,0.9,axis = 0) and t_gr[i] <= np.nanquantile(t_gr,1,axis = 0):
        if wind[i] >= np.nanquantile(wind,0.9,axis = 0) and wind[i] <= np.nanquantile(wind,1,axis = 0):
            w_mn.append(wind[i]) 
            tw.append(t_gr[i])
            
        if ephi[i] >= np.nanquantile(ephi,0.9,axis = 0) and ephi[i] <= np.nanquantile(ephi,1,axis = 0):
            e_mn.append(ephi[i])
            te.append(t_gr[i])
            
        if size[i] >= np.nanquantile(size,0.9,axis = 0) and size[i] <= np.nanquantile(size,1,axis = 0):
            s_mn.append(size[i])
            ts.append(t_gr[i])
            
        if rat[i] >= np.nanquantile(rat,0.9,axis = 0) and rat[i] <= np.nanquantile(rat,1,axis = 0):
            r_mn.append(rat[i])
            tr.append(t_gr[i])
            
        if cenlt[i] >= np.nanquantile(cenlt,0.9,axis = 0) and cenlt[i] <= np.nanquantile(cenlt,1,axis = 0):
            ct_mn.append(cenlt[i])
            tct.append(t_gr[i])
            
        if cenln[i] >= np.nanquantile(cenln,0.9,axis = 0) and cenln[i] <= np.nanquantile(cenln,1,axis = 0):
            cn_mn.append(cenln[i])
            tcn.append(t_gr[i])
            
        if ga[i] >= np.nanquantile(ga,0.9,axis = 0) and ga[i] <= np.nanquantile(ga,1,axis = 0):
            g_mn.append(ga[i])
            tg.append(t_gr[i])
            
        if ef[i] >= np.nanquantile(ef,0.9,axis = 0) and ef[i] <= np.nanquantile(ef,1,axis = 0):
            eh_mn.append(ef[i])
            teh.append(t_gr[i])
       
```

```python
%matplotlib inline
from sklearn.metrics import r2_score
import matplotlib.ticker as mticker

fig = plt.figure(figsize=(20,12),dpi=150)
fig.suptitle("90th Percentile > 10-day Lagged 850-hPa GL Temperature Anomaly and 90th Percentile > Anom Metric") 

#loop through each member
ax1 = plt.subplot(4,2,1)
ax2 = plt.subplot(4,2,2)
ax3 = plt.subplot(4,2,3)
ax4 = plt.subplot(4,2,4)
ax5 = plt.subplot(4,2,5)
ax6 = plt.subplot(4,2,6)
ax7 = plt.subplot(4,2,7)
ax8 = plt.subplot(4,2,8)

z = np.polyfit(tw, w_mn, 1)
y_hat = np.poly1d(z)(tw)
p = np.poly1d(z)
ax1.scatter(tw,w_mn)
ax1.plot(tw,p(tw),"r--")
text = f"$y={z[0]:0.3f}\;x{z[1]:+0.3f}$\n$R^2 = {r2_score(w_mn,y_hat):0.3f}$"
ax1.text(0.05, 0.95, text,transform=ax1.transAxes,
     fontsize=8, verticalalignment='top')
ax1.set_xlabel('Temperature Anom ($^o$)',fontsize = 12)
ax1.set_ylabel('Zonal Mean Zonal Wind (m/s)',fontsize = 12)
ax1.tick_params(axis='both', which='major', labelsize=10);
    
z = np.polyfit(tr, r_mn, 1)
y_hat = np.poly1d(z)(tr)
p = np.poly1d(z)
ax2.scatter(tr,r_mn, color = 'purple')
ax2.plot(tr,p(tr),"r--")
text = f"$y={z[0]:0.3f}\;x{z[1]:+0.3f}$\n$R^2 = {r2_score(r_mn,y_hat):0.3f}$"
ax2.text(0.05, 0.95, text,transform=ax2.transAxes,
     fontsize=8, verticalalignment='top')
ax2.set_xlabel('Temperature Anom ($^o$)',fontsize = 12)
ax2.set_ylabel('Ratio (m)',fontsize = 12)
ax2.tick_params(axis='both', which='major', labelsize=10);
    
z = np.polyfit(te, e_mn, 1)
y_hat = np.poly1d(z)(te)
p = np.poly1d(z)
ax3.scatter(te,e_mn, color = 'red')
ax3.plot(te,p(te),"r--")
text = f"$y={z[0]:0.3f}\;x{z[1]:+0.3f}$\n$R^2 = {r2_score(e_mn,y_hat):0.3f}$"
ax3.text(0.05, 0.95, text,transform=ax3.transAxes,
     fontsize=8, verticalalignment='top')
ax3.set_xlabel('Temperature Anom ($^o$)', fontsize = 12)
ax3.set_ylabel('Rotation Angle ($^o$)', fontsize = 12)
ax3.tick_params(axis='both', which='major', labelsize=10);

z = np.polyfit(ts, s_mn, 1)
y_hat = np.poly1d(z)(ts)
p = np.poly1d(z)
ax4.scatter(ts,s_mn, color = 'magenta')
ax4.plot(ts,p(ts),"r--")
text = f"$y={z[0]:0.3f}\;x{z[1]:+0.3f}$\n$R^2 = {r2_score(s_mn,y_hat):0.3f}$"
ax4.text(0.05, 0.95, text,transform=ax4.transAxes,
     fontsize=8, verticalalignment='top')
ax4.set_xlabel('Temperature Anom ($^o$)', fontsize = 12)
ax4.set_ylabel('Ellipse Area (m$^2$)', fontsize = 12)
ax4.tick_params(axis='both', which='major', labelsize=10);
    
z = np.polyfit(tct, ct_mn, 1)
y_hat = np.poly1d(z)(tct)
p = np.poly1d(z)
ax5.scatter(tct,ct_mn, color = 'green')
ax5.plot(tct,p(tct),"r--")
text = f"$y={z[0]:0.3f}\;x{z[1]:+0.3f}$\n$R^2 = {r2_score(ct_mn,y_hat):0.3f}$"
ax5.text(0.05, 0.95, text,transform=ax5.transAxes,
     fontsize=8, verticalalignment='top')
ax5.set_xlabel('Temperature Anom ($^o$)', fontsize = 12)
ax5.set_ylabel('Central Latitude ($^o$)', fontsize = 12)
ax5.tick_params(axis='both', which='major', labelsize=10);
    
z = np.polyfit(tcn, cn_mn, 1)
y_hat = np.poly1d(z)(tcn)
p = np.poly1d(z)
ax6.scatter(tcn,cn_mn, color = 'cyan')
ax6.plot(tcn,p(tcn),"r--")
text = f"$y={z[0]:0.3f}\;x{z[1]:+0.3f}$\n$R^2 = {r2_score(cn_mn,y_hat):0.3f}$"
ax6.text(0.05, 0.95, text,transform=ax6.transAxes,
    fontsize=8, verticalalignment='top')
ax6.set_xlabel('Temperature Anom ($^o$)', fontsize = 12)
ax6.set_ylabel('Central Longitude ($^o$)', fontsize = 12)
ax6.tick_params(axis='both', which='major', labelsize=10);
    
z = np.polyfit(tg, g_mn, 1)
y_hat = np.poly1d(z)(tg)
p = np.poly1d(z)
ax7.scatter(tg, g_mn, color = 'yellow')
ax7.plot(tg,p(tg),"r--")
text = f"$y={z[0]:0.3f}\;x{z[1]:+0.3f}$\n$R^2 = {r2_score(g_mn,y_hat):0.3f}$"
ax7.text(0.05, 0.95, text,transform=ax7.transAxes,
    fontsize=8, verticalalignment='top')
ax7.set_xlabel('Temperature Anom ($^o$)', fontsize = 12)
ax7.set_ylabel('GPH Anomaly (m)', fontsize = 12)
ax7.tick_params(axis='both', which='major', labelsize=10);
    
z = np.polyfit(teh, eh_mn, 1)
y_hat = np.poly1d(z)(teh)
p = np.poly1d(z)
ax8.scatter(teh,eh_mn, color = 'orange')
ax8.plot(teh,p(teh),"r--")
text = f"$y={z[0]:0.3f}\;x{z[1]:+0.3f}$\n$R^2 = {r2_score(eh_mn,y_hat):0.3f}$"
ax8.text(0.05, 0.95, text,transform= ax8.transAxes,
    fontsize=8, verticalalignment='top')
ax8.set_xlabel('Temperature Anom ($^o$)', fontsize = 12)
ax8.set_ylabel('Eddy Heat Flux (Kg/ms$^2$)', fontsize = 12)
ax8.tick_params(axis='both', which='major', labelsize=10);

plt.tight_layout()
plt.show()
fig.savefig("GLAnom_10MaxMax.png")
```
