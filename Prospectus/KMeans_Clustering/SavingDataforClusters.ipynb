```python
year = [i for i in range(1979,2020,1)]
```

First, let's do geopotential height data at 10hPA, north of 20N latitude. 

```python
#select, slice, and combine the xarrays, removing irrelevant dates

for y in year:
    #print(y)
    if y == 1979:
        gfile = xr.open_dataset("/erai/"+str(y)+"/v."+str(y)+".nc")
        g_files = gfile["v"]
        #g_files
        height1 = g_files.loc[dict(time=slice('1979-11-01','1979-12-31T18:00:00'),lat=slice(0,90),lev=100)]
        #print(height)
        continue 
        
    if 1979 < y < 2015:
        gfile = xr.open_dataset("/erai/"+str(y)+"/v."+str(y)+".nc")
        g_files = gfile["v"]
        #g_files
        
        if y == 1980 or y == 1984 or y == 1988 or y == 1992 or y == 1996 or y == 2000 or y == 2004 or y == 2008 or y == 2012:
            date1 = dt.datetime(y,4,1,0)  #first date to plot
            total_days = 214
            first_fhr = 0
            hours = total_days * 24
            hr_inc = 6
            times = [date1 + dt.timedelta(hours=x) for x in range(0,hours,hr_inc)]
            g_files = g_files.drop_sel(dict(time=times[:]))
            g_data = g_files.loc[dict(lat=slice(0,90),lev=100)]
            height1 = xr.concat([height1, g_data], dim ="time")
            continue
            
        else:
            date1 = dt.datetime(y,4,2,0)  #first date to plot
            total_days = 213
            first_fhr = 0
            hours = total_days * 24
            hr_inc = 6
            times = [date1 + dt.timedelta(hours=x) for x in range(0,hours,hr_inc)]
            g_files = g_files.drop_sel(dict(time=times[:]))
            g_data = g_files.loc[dict(lat=slice(0,90),lev=100)]
            height1 = xr.concat([height1, g_data], dim ="time")
            continue
            
        continue
        
    if 2014 < y < 2019:
        gfile = xr.open_dataset("/erai/"+str(y)+"/v."+str(y)+".nc")
        g_files = gfile["v"]
        if y == 2015:
            date1 = dt.datetime(y,4,2,0)  #first date to plot
            total_days = 213
            first_fhr = 0
            hours = total_days * 24
            hr_inc = 6
            times = [date1 + dt.timedelta(hours=x) for x in range(0,hours,hr_inc)]
            g_files = g_files.drop_sel(dict(time=times[:]))
            height2 = g_files.loc[dict(lat=slice(0,90),lev=100)]
            continue
            
        if y == 2016:
            date1 = dt.datetime(y,4,1,0)  #first date to plot
            total_days = 214
            first_fhr = 0
            hours = total_days * 24
            hr_inc = 6
            times = [date1 + dt.timedelta(hours=x) for x in range(0,hours,hr_inc)]
            g_files = g_files.drop_sel(dict(time=times[:]))
            g_data = g_files.loc[dict(lat=slice(0,90),lev=100)]
            height2 = xr.concat([height2, g_data], dim ="time")
            continue
        else:
            date1 = dt.datetime(y,4,2,0)  #first date to plot
            total_days = 213
            first_fhr = 0
            hours = total_days * 24
            hr_inc = 6
            times = [date1 + dt.timedelta(hours=x) for x in range(0,hours,hr_inc)]
            g_files = g_files.drop_sel(dict(time=times[:]))
            g_data = g_files.loc[dict(lat=slice(0,90),lev=100)]
            height2 = xr.concat([height2, g_data], dim ="time")
            continue
            
        continue
        
    if y == 2019:
        gfile = xr.open_dataset("/erai/"+str(y)+"/v."+str(y)+".nc")
        g_files = gfile["v"]
        #g_files
        date1 = dt.datetime(2019,1,1,0)  #first date to plot
        total_days = 91
        first_fhr = 0
        hours = total_days * 24
        hr_inc = 6
        times = [date1 + dt.timedelta(hours=x) for x in range(0,hours,hr_inc)]
        g_data = g_files.loc[dict(time=times[:],lat=slice(0,90),lev=100)]
        height2 = xr.concat([height2, g_data], dim ="time")
        continue
```

```python
#pickle out the height data just so that I don't need to re-run
pickle.dump(height1, open("V100_79thru.p", 'wb'))
pickle.dump(height2, open("V100_79after.p", 'wb')) 
```
