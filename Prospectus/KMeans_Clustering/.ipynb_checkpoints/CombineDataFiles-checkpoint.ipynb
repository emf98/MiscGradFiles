```python
infile = open("./ThruAfterPickles/50Temp_97thru.p", 'rb')
t1 = pickle.load(infile)
infile.close()

infile = open("./ThruAfterPickles/50Temp_97after.p", 'rb')
t2 = pickle.load(infile)
infile.close()
```

```python
#start with original array
t1

#reweight the latitudes for temp
weights_t=np.cos(np.deg2rad(t1.lat))
weights_t.name = "weights"

#reassign weighted latitudes
t1_new = t1.assign_coords(lat=weights_t.values)

##TEMP
#t1_F = t1_new.stack(latlon = ("lat","lon"))
t1_F = t1_new.drop_vars("lev")
t1_F
```

```python
#DO IT AGAIN FOR THE OTHER DATASETS
t2

#reweight the latitudes for temp
weights_t=np.cos(np.deg2rad(t2.lat))
weights_t.name = "weights"
#print(weights)

#reassign weighted latitudes
t2_new = t2.assign_coords(lat=weights_t.values)

#stack lat and lon together
#t2_F = t2_new.stack(latlon = ("lat","lon"))
t2_F = t2_new.drop_vars("lev")
t2_F
```

```python
#COMBINE TIMES for TEMP
t1_Farray = t1_F.values
t2_Farray = t2_F.values

final_t = np.concatenate((t1_Farray,t2_Farray),axis=0)

corrected_final_t = np.where(final_t > 1000., final_t/100., final_t)
#final_h
```

```python
##TEMP
t_time1_coord = t1_F.time.values
t_time2_coord = t2_F.time.values

#t_final_latlon = t1_F.latlon.values

t_time_final =  np.concatenate((t_time1_coord,t_time2_coord),axis=0)
```
