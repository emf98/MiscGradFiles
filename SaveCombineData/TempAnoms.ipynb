```python
#import statements... I think these are all of the relevant ones to what I am doing here. 
import datetime as dt
import numpy as np
import xarray as xr
import pickle
```

```python
infile = open("FullNH_1000_temp_1959_PT1.p", 'rb') 
file1 = pickle.load(infile)
#test = pickle.load(infile)
infile.close()

infile = open("FullNH_1000_temp_1959_PT2.p", 'rb') 
file2 = pickle.load(infile)
infile.close()
```

```python
##save year index
year = [i for i in range(1959,2022)] #indicate years for data, needed for opening files
```

```python
##save lat lon indices
lats = []
lons = []
testfile = xr.open_dataset("../era5/gph/era5_gph_1979.nc")
lats.append(testfile.latitude.values)
lons.append(testfile.longitude.values)

lats = lats[0]
lons = lons[0]
```

```python
##save day indices
import pandas as pd

s = pd.date_range('2011-11-01', periods=152, freq='D').to_series()
test1 = s.dt.strftime('%m-%d')
times = []
for i in range(len(s)):
    times.append(test1[i])
```

```python
temp = np.concatenate((file1,file2), axis=0)
temp.shape
```

```python
##create array for gph ... 
t = xr.DataArray(data= temp, dims = ["year","day","lat","lon"],coords = dict(year = year,
                                                                                day = times,
                                                                                lat = lats,
                                                                                lon = lons))
t
```

```python
#select cap
cap_t = t.loc[dict(lat=slice(90,60))]
cap_t
```

```python
#select europe
e_t = t.loc[dict(lat=slice(70,40),lon=slice(-5,30))]
e_t
```

```python
#select great lakes 
gl_t = t.loc[dict(lat=slice(50,40),lon=slice(-90,-80))]
gl_t
```

```python
##pickle out these files
pickle.dump(cap_t.values, open("1959cap_temp1000.p", 'wb'))
pickle.dump(e_t.values, open("1959europe_temp1000.p", 'wb'))
pickle.dump(gl_t.values, open("1959gl_temp1000.p", 'wb'))
```

```python
#create weights
weights=np.cos(np.deg2rad(cap_t.lat))
cap_t_weight = cap_t.weighted(weights).mean(dim="lon").mean(dim="lat") #average over the full area
cap_t_weight
```

```python
##calculate daily mean anomaly
##cap
dailymean_cap = np.mean(cap_t_weight,axis=0)
anom_cap = cap_t_weight - dailymean_cap
anom_cap.shape
```

```python
anom_cap
```

```python
pickle.dump(anom_cap.values, open("1959weightedANOMcap_temp1000.p", 'wb'))
```
